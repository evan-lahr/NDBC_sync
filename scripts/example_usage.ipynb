# Example: ndbc_sync compiles ~30 years of data
#### data from https://www.ndbc.noaa.gov/station_history.php?station=46029

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import ndbc_tools

n46029_all=ndbc_tools.ndbc_sync('../data/*.txt')
n46029_all
```

# Plot concatenated data

```python
df=n46029_all

fig, axs = plt.subplots(5, 1, sharex=True)

axs[0].scatter(df['datetime'],df['BAR'],marker='.',s=4,c='k')
axs[1].scatter(df['datetime'],df['WSPD'],marker='.',s=4,c='k')
axs[2].scatter(df['datetime'],df['WVHT'],marker='.',s=4,c='k')
axs[3].scatter(df['datetime'],df['MWD'],marker='.',s=4,c='k')
axs[4].scatter(df['datetime'],df['APD'],marker='.',s=4,c='k')

axs[0].set_ylabel('BAR')
axs[1].set_ylabel('WSPD')
axs[2].set_ylabel('WVHT')
axs[3].set_ylabel('MWD')
axs[4].set_ylabel('APD')

axs[4].set_axisbelow(True)

plt.subplots_adjust(wspace=0.1, hspace=0.1)
fig.autofmt_xdate()
fig.set_size_inches(15,6)
```
